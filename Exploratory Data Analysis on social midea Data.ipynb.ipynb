```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.select_dtypes(include=['int64', 'float64']).corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python

sns.histplot(df['Platform'])
plt.show()
```

```python
sns.countplot(x = df['Hour'],palette=['blue', 'red', 'yellow'])
plt.legend(title="Hour", labels=df['Hour'].unique())
plt.show()
```

```python

plt.pie(df['Sentiment'].value_counts(),startangle=90,counterclock=False,wedgeprops={'edgecolor':'black'})
plt.title("Platform")

plt.show()
```

```python
plt.bar(df['Platform'], df['Retweets'])  
plt.xlabel('Hashtags')
plt.ylabel('Country')
plt.title('Hour')
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.bar(df['Platform'], df['Retweets'])  
plt.xlabel('Status Mode')
plt.ylabel('Retweets')
plt.title('Hour')
plt.xticks(rotation=45) 
plt.show()
```

```python

yearly_sales = df.groupby('Year')['Hour'].sum().reset_index()
sns.lineplot(x=yearly_sales['Year'], y=yearly_sales['Hour'], marker='o', color='blue')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Day')
plt.title('Month')
plt.show()
```

```python

status_quantity = df.groupby('Platform')['Hour'].sum()
plt.bar(status_quantity.index, status_quantity.values)
plt.xlabel('Country')
plt.ylabel('month')
plt.title('Sentiment')
plt.xticks(rotation=45) 
plt.show()
```

```python
status_dealsize_sales = df.groupby(['Platform', 'Year'])['Month'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Platform', y='Month', hue='Year', data=status_dealsize_sales, palette=['blue','red','orange'])

plt.xlabel('Country')
plt.ylabel('Day')
plt.title('Retweets')
plt.xticks(rotation=45)
plt.legend(title='likes')
plt.show()
```

```python

status_dealsize_QuantityOrder = df.groupby(['Platform', 'Year'])['Hour'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Platform', y='Hour', hue='Year', data=status_dealsize_QuantityOrder, palette=['pink','yellow','skyblue'])

plt.xlabel('Country')
plt.ylabel('Year')
plt.title('Retweets')
plt.xticks(rotation=45)
plt.legend(title='Likes')
plt.show()
```

```python

top_states = df.groupby('Country')['Year'].sum().nlargest(10).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='Country', y='Year', data=top_states, palette='viridis')

plt.xlabel('Country')
plt.ylabel('Year')
plt.title('Top 10 Countries with Highest User')
plt.xticks(rotation=45)  
plt.show()
```

```python
top_countries = df.groupby('Country')['Hour'].sum().nlargest(10)

plt.figure(figsize=(12, 6))
plt.bar(top_countries.index, top_countries.values, color='royalblue')

plt.xlabel('Country')
plt.ylabel('Month')
plt.title('Top 10 Countries with Highest Hashtags')
plt.xticks(rotation=45)  

plt.show()
```

```python

df.groupby('Year')['Country'].sum().plot(kind='pie',autopct='%1.1f%%', title="Sales Distribution by Territory Region")
plt.ylabel('')
plt.show()
```
